# Guided Project: Star Wars Survey

```python
import pandas as pd
import numpy as np
star_wars = pd.read_csv('star_wars.csv', encoding='ISO-8859-1')
```

```python
star_wars.head(10)
```

```python
star_wars.columns
```

## Removing Null Value Respondent ID Rows

```python
#Find null values
star_wars.isnull().sum()
```

```python
#Select finite values using notnull()
star_wars = star_wars[pd.notnull(star_wars['RespondentID'])]
star_wars.isnull().sum()
```

## Converting 2 Yes/No Columns To Booleans

```python
yes_no = {"Yes": True, "No": False}
star_wars['Have you seen any of the 6 films in the Star Wars franchise?'] = star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].map(yes_no)
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'] = star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].map(yes_no)
star_wars.head(5)
```

```python
print(star_wars[3:9].head(2))
```

```python
film_names = {"Star Wars: Episode I  The Phantom Menace": True,
              np.nan: False,
"Star Wars: Episode II  Attack of the Clones": True,
"Star Wars: Episode III  Revenge of the Sith": True,
"Star Wars: Episode IV  A New Hope": True, 
"Star Wars: Episode V The Empire Strikes Back": True,
"Star Wars: Episode VI Return of the Jedi": True}

for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(film_names)
star_wars.head(2)
```

## Renaming Columns

```python
star_wars = star_wars.rename(columns={'Which of the following Star Wars films have you seen? Please select all that apply.': 'seen_1', 'Unnamed: 4': 'seen_2', 'Unnamed: 5': 'seen_3', 'Unnamed: 6': 'seen_4', 'Unnamed: 7': 'seen_5', 'Unnamed: 8': 'seen_6'})
star_wars.columns
```

## Converting String Columns to Numeric Values

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
```

```python
star_wars = star_wars.rename(columns={'Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.': "ranking_1",
       'Unnamed: 10': "ranking_2", 'Unnamed: 11': 'ranking_3', 'Unnamed: 12': 'ranking_4', 'Unnamed: 13': 'ranking_5',
       'Unnamed: 14': 'ranking_6'})
star_wars.columns
```

```python
star_wars[star_wars.columns[9:15]].mean()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.bar(range(6), star_wars[star_wars.columns[9:15]].mean())
```

The graph shows that respondents ranked the 4th Star Wars film as the their favourite, with their least favourite film being the 3rd (and final to be released) film. This may also down to the famous scenes in which the death star is destroyed. 

```python
star_wars[star_wars.columns[3:9]].sum()
```

```python
plt.bar(range(6), star_wars[star_wars.columns[3:9]].sum())
```

The most viewed chart follows an almost identical order to the rankings.

## Exploring the Data By Gender

```python
males = star_wars[star_wars["Gender"] == "Male"]
females = star_wars[star_wars["Gender"] == "Female"]
```

## Ranking By Gender (Male/Female)

```python
plt.bar(range(6), males[males.columns[9:15]].mean())
plt.show()
plt.bar(range(6), females[females.columns[9:15]].mean())
plt.show()
```

Male respondents prefer episode 4 and 6 equally whereas female respondents prefer episode 6 to episode 4.

## View Count By Gender (Male/Female)

```python
plt.bar(range(6), males[males.columns[3:9]].sum())
plt.show()
plt.bar(range(6), females[females.columns[3:9]].sum())
plt.show()
```

## Findings

Based on view count, the Star Wars films are more popular with males than with females. The popularity of each film based on views follows an identical order.
